# 📊 Overview of Databricks SQL (DBSQL)

Databricks SQL is a powerful environment for data analysis, querying, and visualization. It enables analysts, engineers, and business users to interact with data using SQL and build dashboards for insights and monitoring.

---

## 👤 Switching to SQL Persona

To access Databricks SQL features, users must switch to the **SQL persona** in the workspace. This persona provides access to:

- **SQL Editor** – Write and execute SQL queries.
- **Queries** – Manage saved and scheduled queries.
- **Dashboards** – Create visual reports and interactive views.
- **Alerts** – Set up conditional notifications based on query results.

---

## ⚙️ SQL Warehouses

SQL Warehouses are the **compute resources** that power SQL queries in Databricks.

### 🔧 Creating a SQL Warehouse
- Users are guided through the creation of a SQL warehouse (e.g., "Demo Warehouse").
- Configuration settings are reviewed, including scaling options and auto-stop settings.

---

## 📈 Creating a Dashboard

A hands-on demonstration walks through importing a sample dataset (e.g., NYC taxi trips) and visualizing the data.

### 💡 Dashboard Features:
- Refresh dashboard data on demand.
- View the underlying SQL for each visualization.
- Understand table sources using the three-tier namespace: `catalog.schema.table`.

### 🎨 Modifying Visualizations:
- Edit existing visuals.
- Add new visual types such as pie charts or bar graphs.
- Customize filters and parameters to refine insights.

---

## 🤝 Sharing Dashboards

- Dashboards can be shared with collaborators across your organization.
- Permissions can be configured to control **view** or **edit** access for other users.

---

## ✍️ Writing Custom SQL Queries

### 🔌 Connecting to SQL Warehouse:
- Users connect to their created warehouse and choose a database for analysis.

### 🔎 Exploring and Querying Data:
- Preview table schemas to understand available columns.
- Use standard SQL syntax, including:
  - `GROUP BY`
  - Aggregation functions (`SUM`, `COUNT`, `AVG`, etc.)

### 💾 Saving Queries:
- Save frequently used queries.
- Pin query results directly to dashboards for dynamic visual updates.

---

## 🔁 Scheduling Refreshes

- Queries and dashboards can be configured to **auto-refresh** on a schedule.
- Ensures up-to-date data is always available for decision-making.

---

## 🚨 Setting Up Alerts

- Alerts can be triggered based on conditions from saved queries.
- Email or workspace notifications keep users informed when data meets specified thresholds.

---

## ✅ Conclusion

This session provides a full overview of Databricks SQL — from creating warehouses to building dashboards and setting alerts.  
Users gain practical skills to analyze data, create visuals, and share insights effectively, all within a powerful, SQL-first environment.
